Input Declarations

```python
n = int(input("Enter no of states: "))
prob = {}
state_name = {}
for i in range(n):
    prob[i] = float(input("Initial probability for state: "))
    s = input("State Name: ")
    state_name[i] = s
print(prob)
```

Input - Emission states and Probability

```python
op = input("Emission States: ")
print("Output Emissions: ",output)
op_emit = op.split()
p = n*(len(op_emit))
print(f"Possible Emission: {p}")

emit = {}
for i in range(n):
    for j in op_emit:
        emit[i,j] = float(input("Emission Probability for sequence i to j: "))
print("\n",emit)
```

Transmission Probability

```python
t = n**2
print(f"No. of transmissions for given state: {t}")
transition = {}
for i in range(n):
    for j in range(n):
        transition[i,j] = float(input("Transmission Probability for sequence i to j: "))
print("\n",transition)
```

Alpha and Beta values

```python
alpha = [[],]
beta = [[]]
for i in range(n):
    alpha[0].extend([prob[i]])
    beta[0].extend([1])

print(alpha[0])
print(beta[0])
```

Gama Declarations

```python
gamma = []
```

Forward Procedure

```python
def Trellis_FD(n, out, transition, emit, alpha):
    alp = []
    lists = []
    # i --> current state(time t+1)
    # j --> Previous State(time t)
    for i in range(n):
        for j in range(n):
            # This inner loop calculates n products for alpha values of each state at time t+1 
            #(For each iteration of i, alpha value for each state at time t+1 are calculated using alpha values at time t)
            value = alpha[j]*emit[j,out]*transition[j,i]
            #Each product is appended to 'lists' (total n such products will be stored for each iteration of i)
            lists.append(value)
        else:
            if len(lists)==2:
                #In case of 2 states, we only get 2 products to add for getting alpha value at time t+1. Store that sum in another list 'alp'
                alp.append(lists[0]+lists[1])
                # Empty 'lists' for using it for next iteration of i
                lists=[]
            elif len(lists)==3:
                #In case of 3 states, we only get 3 products to add for getting alpha value at time t+1.Store that sum in another list 'alp'
                alp.append(lists[0]+lists[1]+lists[2])
                # Empty 'lists' for using it for next iteration of i
                lists=[]
    return alp # Return list of alpha values of current state (at time t+1) 
```

Backward Procedure

```python
def Trellis_BD(n, out, transition, emit, beta):
    lists = []
    beta1 = []
    # i --> Current State(time t)
    # j --> Future State(time t+1)
    for i in range(n):
        for j in range(n):
            # This inner loop calculates n products for beta values of each state at time t 
            #(For each iteration of i, beta value for each state at time t are calculated using beta values at time t+1)
            value = beta[j]*emit[i,out]*transition[i,j]
            #Each product is appended to 'lists' (total n such products will be stored for each iteration of i)
            lists.append(value)
        else:
            if len(lists)==2:
                #In case of 2 states, we only get 2 products to add for getting beta value at time t. Store that sum in another list 'beta1'
                beta1.append(lists[0]+lists[1])
                # Empty 'lists' for using it for next iteration of i
                lists = []
            elif len(lists)==3:
                #In case of 3 states, we only get 3 products to add for getting beta value at time t. Store that sum in another list 'beta1'
                beta1.append(lists[0]+lists[1]+lists[2])
                # Empty 'lists' for using it for next iteration of i
                lists=[]
    return beta1  # Return list of beta values of current state (at time t) 
```

Trellis Forward Procedure

```python
c=0
for i in out:
    new = Trellis_FD(n,i,transition,emit,alpha[c])
    alpha.append(new)
    # Storing all the lists returned by Trellis_FD function into list 'alpha'
    c+=1
print(alpha) 
```

Reverse List

```python
def rev(lis):
    new_lst = lis[::-1]
    return new_lst
out = rev(out)
print(out)
```

Trellis Backward Procedure

```python
c=0
for i in out:
    new = Trellis_BD(n,i,transition,emit,beta[c])
    beta.append(new)
    # Storing all the lists returned by Trellis_FD function into list 'alpha'
    c+=1
print(beta)
```

Probability of Forward and Backward Sequence

```python
a = alpha[-1]
prob_fd_seq=0
prob_bd_seq=0
for i in a:
    prob_fd_seq+=i
b = beta[-1]
c = alpha[0]
for i in range(len(b)):
    prob_bd_seq+=(b[i]*c[i])
print(f"\n{prob_fd_seq}\n{prob_bd_seq}")
```

Gamma value calculation

```python
def best_seq(n,alpha,beta):
    if n==2:
        num1 = alpha[0]*beta[0]
        num2 = alpha[1]*beta[1]
        total = num1+num2
        gamma1 = num1/total
        gamma2 = num2/total
        return [gamma1,gamma2]
    elif n==3:
        num1 = alpha[0]*beta[0]
        num2 = alpha[1]*beta[1]
        num3 = alpha[2]*alpha[2]
        total = num1+num2+num3
        gamma1 = num1/total
        gamma2 = num2/total
        gamma3 = num3/total
        return [gamma1,gamma2,gamma3]
```

```python
print(alpha)
beta = beta[::-1]
print(beta)
```

Find best squence

```python
counter1 = 0
counter2 = 0
state0 = []
state1 = []
state2 = []
gamma = []
while counter1 < len(alpha) and counter2 < len(beta):
    lists = []
    lists = best_seq(n,alpha[counter1],beta[counter2])
    gamma.append(lists)
    if n==2:
        if lists[0]>=lists[1]:
            state0.append(lists[0])
            state1.append(0)
        elif lists[0]<lists[1]:
            state1.append(lists[1])
            state0.append(0)
    elif n==3:
        if lists[0]>lists[1] and lists[0]>lists[2]:
            state0.append(lists[0])
            state1.append(0)
            state2.append(0)
        elif lists[0]<lists[1] and lists[2]<lists[1]:
            state1.append(lists[0])
            state0.append(0)
            state2.append(0)
        elif lists[0]<lists[2] and lists[1]<lists[2]:
            state2.append(lists[0])
            state0.append(0)
            state1.append(0)
        elif lists[0]==lists[1] and lists[1]==lists[2]:
            state0.append(lists[0])
            state1.append(0)
            state2.append(0)
    counter1+=1
    counter2+=1
print(f"Gamma values: {gamma}\n{len(gamma)}")
print(f"State0: {state0}\nState1: {state1}\nState2: {state2}")
```

Display the sequence

```python
j=0
n1 = n
state = []
while(j<len(gamma)):
    if n1==2:
        if state0[j]>state1[j]:
            state.append(0)
        else:
            state.append(1)
    elif n1==3:
        if state0[j]>state1[j] and state0[j]>state2[j]:
            state.append(0)
        elif state1[j]>state0[j] and state1[j]>state2[j]:
            state.append(1)
        elif state2[j]>state0[j] and state2[j]>state1[j]:
            state.append(2)
    j+=1
print(state)
state_display = []
for k in state:
    if k in state_name.keys():
        state_display.append(state_name[k])
print(f"\n{state_display}")
```
