```python
import gensim
from gensim.models import Word2Vec
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# New dataset: Customer Reviews (Expanded Sentences)
sentences = [
    "This phone has an amazing battery life.",
    "I love the camera quality of this phone.",
    "The battery life of this phone is terrible.",
    "Great performance and smooth user experience.",
    "This laptop is lightweight and easy to carry.",
    "I hate how slow this laptop runs after an update.",
    "The display resolution is crystal clear.",
    "Worst laptop I've ever used, very disappointing.",
    "Highly recommend this product to everyone!",
    "Not worth the price, totally disappointed."
]

# Tokenize the sentences (convert to lowercase and split into words)
tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in sentences]

# Train the Word2Vec model
model_w2v = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=10, min_count=1, workers=5)

# Test word vectors (semantic similarity)
word = 'battery'
similar_words = model_w2v.wv.most_similar(word, topn=10)
print(f"\n Words most similar to '{word}':")
for sim_word, sim_score in similar_words:
    print(f"{sim_word}: {sim_score:.4f}")

# Save the model for future use
model_w2v.save("word2vec_reviews.bin")

# Load and test the saved model
loaded_model = Word2Vec.load("word2vec_reviews.bin")
print("\n Word2Vec model loaded successfully!")
```
