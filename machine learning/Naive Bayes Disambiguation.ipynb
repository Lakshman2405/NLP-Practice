```python
import pandas as pd
import nltk

df = pd.read_csv('Bank.csv')
df.info()
df.columns.to_list

```

```python
df.head()
print("Current columns:", df.columns.tolist())

df.drop('Unnamed: 0', inplace=True, axis=1, errors='ignore')
```

```python
df.value_counts('Sense of the word "Bank"')
```

```python
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize,word_tokenize
df_finance = df[df['Sense of the word "Bank"'] == 'Financial Institution']
df_river = df[df['Sense of the word "Bank"'] == 'River Border']

stopword = set(stopwords.words('english'))
def preprocess(text):
    data = sent_tokenize(text)
    data = [word_tokenize(word) for word in data ]
    data = [word for i in data for word in i if word.lower() not in stopword]
    return data

token_fin = [preprocess(data) for data in df_finance['Sentences for Training ']]
token_fin = [word for i in token_fin for word in i]
count_fin = Counter(token_fin)

token_riv = [preprocess(data) for data in df_river['Sentences for Training ']]
token_riv = [word for i in token_riv for word in i]
count_riv = Counter(token_riv)

#count_fin
```

```python
import numpy as np
df_find = df[df['Sense of the word "Bank"'] == '?']

tot = df.value_counts('Sense of the word "Bank"')['River Border'] + df.value_counts('Sense of the word "Bank"')['Financial Institution']
p_riv = df_river.shape[0]/tot
p_fin = df_finance.shape[0]/tot

doc = []
for ind,val in df_find.iterrows():
    doc.append(val['Sentences for Training '])

p_finance = np.log2(p_fin)
p_river = np.log2(p_riv)
for data in doc:
    data = word_tokenize(data)
    for i in data:
        p_finance += np.log2((count_fin.get(i,0)+1)/(sum(count_fin.values())+len(set(count_fin.keys()))))
        p_river += np.log2((count_riv.get(i,0)+1)/(sum(count_riv.values())+len(set(count_riv.keys()))))
    print(data)
    if p_finance >= p_river:
        print("data belongs to finance")
    else:
        print("data belongs to river")
    p_finance = np.log2(p_fin)
    p_river = np.log2(p_riv)
```
